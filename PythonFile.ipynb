```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
data = pd.read_csv('/content/Dataset1.csv')

# Filter out rows with unknown values in 'weather_conditions' field
filtered_data = data[data['weather_conditions'] != 'Unknown']

# Group the filtered data
heatmap_data = filtered_data.groupby(['speed_limit', 'weather_conditions']).size().reset_index(name='accident_count')

# Pivot the DataFrame
heatmap_matrix = heatmap_data.pivot(index='speed_limit', columns='weather_conditions', values='accident_count')

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_matrix, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Road Traffic Accidents by Speed Limit and Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Speed Limit')
plt.show()
```
